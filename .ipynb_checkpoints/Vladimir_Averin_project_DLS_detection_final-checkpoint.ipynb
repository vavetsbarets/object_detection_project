```python
# Choosing SSD from Nvidia, importing torch and loading the model
# Use GPU runtime to download the model!!!!
import torch 

precision = 'fp32'
ssd_model = torch.hub.load('NVIDIA/DeepLearningExamples:torchhub', 'nvidia_ssd', model_path = precision)
```

```python
# utilites which helps with preprocessing of images, before going to SSD model!!!!
utils = torch.hub.load('NVIDIA/DeepLearningExamples:torchhub', 'nvidia_ssd_processing_utils')
```

```python
# putting the model to GPU and setting to the evaluation mode
ssd_model.to('cuda')
ssd_model.eval()
```

```python
# some sample photos, from seminar
# tried to include some other photos, but there were errors!!!
photos = [
    'http://images.cocodataset.org/val2017/000000397133.jpg',
    'http://images.cocodataset.org/val2017/000000037777.jpg',
    'http://images.cocodataset.org/val2017/000000252219.jpg',
]
```

```python
# preprocessing from photo to tensor
inputs = [utils.prepare_input(p) for p in photos]
tensor = utils.prepare_tensor(inputs, precision == 'fp16')
```

```python
# making a batch from the tensor
with torch.no_grad():
    detections_batch = ssd_model(tensor)

# boxes per image, filtered by IOU threshold!!!
results_per_input = utils.decode_results(detections_batch)

# Choosing only boxes where classification threshold is beaten
best_results_per_input = [utils.pick_best(results, 0.20) for results in results_per_input]

# finding class labels in the dictionary of the dataset!!!
classes_to_labels = utils.get_coco_object_dictionary()

# visualization of photo with boxes which detect certain classes!!!!
from matplotlib import pyplot as plt
import matplotlib.patches as patches

for image_idx in range(len(best_results_per_input)):
    fig, ax = plt.subplots(1)
    # Show original, denormalized image...
    image = inputs[image_idx] / 2 + 0.5
    ax.imshow(image)
    # ...with detections
    bboxes, classes, confidences = best_results_per_input[image_idx]
    for idx in range(len(bboxes)):
        left, bot, right, top = bboxes[idx]
        x, y, w, h = [val * 300 for val in [left, bot, right - left, top - bot]]
        rect = patches.Rectangle((x, y), w, h, linewidth=1, edgecolor='r', facecolor='none')
        ax.add_patch(rect)
        ax.text(x, y, "{} {:.0f}%".format(classes_to_labels[classes[idx] - 1], confidences[idx]*100), bbox=dict(facecolor='white', alpha=0.5))
plt.show()
```

```python
# function which returns dataset (list) with raw images + boxes location + labels + number of boxes (input: path)!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!

def dir_to_dataset(dir):  
    # arakul: https://pytorch.org/tutorials/beginner/data_loading_tutorial.html !!!!!!!!!!!!!!!!!!!!!!!!!!!
    # reserve arakul: https://www.kaggle.com/yerramvarun/fine-tuning-faster-rcnn-using-pytorch/notebook#Visualization !!!!!!!!!!!!!!!!!!!!!!!!
    list_image_boxes = []
    # FINDING MEANS PER EACH CHANNEL FOR NORMALIZATION!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    # (ESTIMATING ON TRAIN SET ONLY, SAME FOR STD)!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    red_channel_mean = []
    green_channel_mean = []
    blue_channel_mean = []
    # FINDING PROXY OF STD PER EACH CHANNEL (CONSISTENT ESTIMATOR - AVERAGE OF STDS PER EACH PHOTO PER CHANNEL)!!!!!!!!!!!!!!!
    red_channel_std = []
    green_channel_std = []
    blue_channel_std = []

    for i in list(dir.rglob('*.jpg')):
        image = cv2.imread(str(i.resolve()))
        # converting all images to RGB format and to float!!!!!
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB).astype(np.float32)
        # dividing image by 256!!!!! 
        image = image / 255.0

        # average value of each channel per photo!!!!!!
        red_channel_mean.append(image[:, :, 0].mean())
        green_channel_mean.append(image[:, :, 1].mean())
        blue_channel_mean.append(image[:, :, 2].mean())
        # std of each channel per photo!!!!!!!!!!!!!!!!!!!!!
        red_channel_std.append(image[:, :, 0].std())
        green_channel_std.append(image[:, :, 1].std())
        blue_channel_std.append(image[:, :, 2].std())
        
        file_xml = '{}.xml'.format(str(i.resolve())[:-4])
        bbox_raw = []
        label = []
        for i in et.parse(file_xml).getroot().findall('object'):
            # SO INDEX OF THE ELEMENT IN CLASSES LIST IS IT'S LABEL!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
            label.append(classes.index(i.find('name').text))
            # THIS FORMAT, SO THAT IT WILL BE CONVINIENT LATER ON!!!!!!!! 
                # FUNCTIONS TO CONVERT TO CXYWH AND GCXYWH ARE TO BE INTRODUCED LATER ON!!!!!!!!!!!!!!!!!!!!!!!!!!
            bbox_raw.append(int(i.find('bndbox').find('xmin').text))
            bbox_raw.append(int(i.find('bndbox').find('ymin').text))
            bbox_raw.append(int(i.find('bndbox').find('xmax').text))
            bbox_raw.append(int(i.find('bndbox').find('ymax').text))
        bbox_raw = np.asarray(bbox_raw)
        label = np.array(label)
        bbox_raw = bbox_raw.astype('float').reshape(-1, 4)
        label = label.astype('float').reshape(-1, 1)
        # ADDING THE NUMBER OF THE BOXES TO BE USED IN THE COLLATE_FN!!!!!
        num_b = np.asarray(len(label))
        sample = {'image': image, 'bboxes': bbox_raw, 'label': label, 'num_b': num_b}

        list_image_boxes.append(sample)
    
    # MEAN AND STD OF THE TRAIN SET, PER CHANNEL (CONSISTENT ESTIMATOR OF STD)!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    red_channel_mean = np.array(red_channel_mean).mean()
    green_channel_mean = np.array(green_channel_mean).mean()
    blue_channel_mean = np.array(blue_channel_mean).mean()

    red_channel_std = np.array(red_channel_std).mean()
    green_channel_std = np.array(green_channel_std).mean()
    blue_channel_std = np.array(blue_channel_std).mean()

    mean_per_channel = [red_channel_mean, green_channel_mean, blue_channel_mean]
    std_per_channel = [red_channel_std, green_channel_std, blue_channel_std]

    return list_image_boxes, mean_per_channel, std_per_channel
```

```python
# CREATING SPECIFIC RESCALE AND TOTENSOR: SO THAT ALL PHOTOS ARE 3X300X300!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
# MAKE RESCALE SO THAT 300X300 STRAIGHT AWAY, SINCE RARELY PANORAMA!!!! BUT IF RECROP THEN DEFINITELY CUT SOME FRUIT, SO I DECIDED NOT TO DO ANY RECROP!!!!!!!!!!!!!!!!!!!!!!
class Rescale(object):
    """Rescale the image in a sample to a given size.

    Args:
        output_size (tuple or int): Desired output size. If tuple, output is
            matched to output_size. If int, smaller of image edges is matched
            to output_size keeping aspect ratio the same.
    """

    def __init__(self, output_size):
        assert isinstance(output_size, (int, tuple))
        self.output_size = output_size

    def __call__(self, sample):
        image, bboxes, label, num_b = sample['image'], sample['bboxes'], sample['label'], sample['num_b']

        h, w = image.shape[:2]
        if isinstance(self.output_size, int):
            # MAKE RESCALE SO THAT 300X300 STRAIGHT AWAY, SINCE RARELY PANORAMA!!!! BUT IF RECROP THEN DEFINITELY CUT SOME FRUIT
            if h > w:
                # new_h, new_w = self.output_size * h / w, self.output_size
                new_h, new_w = self.output_size, self.output_size
            else:
                # new_h, new_w = self.output_size, self.output_size * w / h
                new_h, new_w = self.output_size, self.output_size
        else:
            new_h, new_w = self.output_size, self.output_size

        new_h, new_w = int(new_h), int(new_w)

        img = transform.resize(image, (new_h, new_w))
        # MAKE RESCALE SO THAT 300X300 STRAIGHT AWAY, SINCE RARELY PANORAMA!!!! BUT IF RECROP THEN DEFINITELY CUT SOME FRUIT
        # ALSO LET'S DIVIDE HERE BY 300 TO NORMALIZE !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
        # FOR Y USED NORMALIZED HEIGHT, FOR X USED NORMALIZED WIDTH!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
        bboxes = bboxes * [new_w / w / 300, new_h / h / 300, new_w / w / 300, new_h / h / 300]

        # CONVERT TO CXCYWH FORMAT!!!!!!!! UPD: THIS WILL BE IMPLEMENTED FURTHER ON AS A SEPARATE FUNCTIONS!!!!!!!!!!
        #bboxes_new = np.zeros((bboxes.shape[0], bboxes.shape[1]))
        #cx and cy
        #bboxes_new[:, 0] = (bboxes[:, 0] + bboxes[:, 1]) / 2
        #bboxes_new[:, 1] = (bboxes[:, 2] + bboxes[:, 3]) / 2
        #w and h
        #bboxes_new[:, 2] = bboxes[:, 1] - bboxes[:, 0]
        #bboxes_new[:, 3] = bboxes[:, 3] - bboxes[:, 2]

        return {'image': img, 'bboxes': bboxes, 'label': label, 'num_b': num_b}


class ToTensor(object):
    """Convert ndarrays in sample to Tensors."""

    def __call__(self, sample):
        image, bboxes, label, num_b = sample['image'], sample['bboxes'], sample['label'], sample['num_b']

        # swap color axis because
        # numpy image: H x W x C
        # torch image: C x H x W
        image = image.transpose((2, 0, 1))
        return {'image': torch.from_numpy(image),
                'bboxes': torch.from_numpy(bboxes), 
                'label': torch.from_numpy(label), 
                'num_b': torch.from_numpy(num_b)}
```

```python
# collate_fn to stack different sizes of tensors (different pictures have different number of boxes!!!)!!!!!!!!!!!!!!!!!!!!!!!!!
def collate_fn(data):
    max_box_num = 1
    # finding max number of boxes on a single photo!!!!!!!!!!!!!!!!!!!!!
    for i in range(len(data)):
        if data[i]['num_b'] > max_box_num:
            max_box_num = data[i]['num_b']

    # inserting redundant values, so that shapes of samples will be the same, so we can use batching!!!!!!!!! 
    for i in range(len(data)):
        # filling boxes space with zeroes, to make the same shape for each sample!!!!!!
        j, k = data[i]['bboxes'].size(0), data[i]['bboxes'].size(1)
        data[i]['bboxes'] = torch.cat([data[i]['bboxes'], torch.zeros((max_box_num - j, k))])
        # filling labels space with number higher than the number of classes, to make the same shape for each sample!!!
        data[i]['label'] = torch.cat([data[i]['label'], 5 * torch.ones((max_box_num - j, 1))])

    return data
```

```python
# setting batch size !!!!!!!!
BATCH_SIZE = 8
```

```python
# TURNING DATASET TO DATALODER!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
def to_DataLoader(dataset, BATCH_SIZE = BATCH_SIZE):
    scale = Rescale(300)
    to_tensor = ToTensor()
    composed = transforms.Compose([scale, to_tensor])
    tensored_scaled_img_boxes = []

    for sample in dataset:
        tensored_scaled_img_boxes.append(composed(sample))

    # I COULD HAVE USED THIS FUNCTION AS A PARAMETER OF THE FUNCTION BELOW, BUT I FOUND IT MORE CONVENIENT TO 
        # USE THIS FUNCTION DIRECTLY HERE!!!!! OUTPUT IN THIS CASE IS MORE PLAUSIBLE FOR ME (IF THERE WERE MORE PHOTOS IN THE DATA, 
        # I WOULD USE COLLATE_FN IN THE DATALOADER FUNCTION BELOW)!!!!!
    data_to_batch = collate_fn(tensored_scaled_img_boxes)
    
    data_loader = torch.utils.data.DataLoader(
        tensored_scaled_img_boxes, batch_size = BATCH_SIZE, shuffle=True, num_workers = 2
    )

    return data_loader
```

### SO THIS FUNCTION WILL BE USED IN THE TRAIN FUNCTION IN THE NEXT PART OF THE PROJECT!!! 
### ALSO IT IS USED FOR THE SANITY CHECK IN THIS PART OF THE PROJECT, WHICH SHOWS THAT BOXES AND PHOTOS ARE CORRECTLY CONNECTED AFTER AUGMENTATION!!!!!!!

```python
def general_augmentation(image, bboxes, num_objects, prob_horizontal_flip = 0.5, prob_vertical_flip = 0.5):
    
    # VERTICAL FLIP WITH PROBABILITY OF 0.5
    # SETTING PROBABILITY OF THE VERTICAL FLIP EXTERNALLY, BECAUSE OF BOXES!!!!!!
    if np.random.random() < prob_horizontal_flip:
        transform1 = transforms.Compose([
                        # first ToPILImage and then ToTensor before normalize!!!!!!!!!!!!!!!!!!!!!!!!
                        transforms.ToPILImage(),
                        transforms.RandomVerticalFlip(p = 1),
                        transforms.ToTensor()
                    ])
        image = transform1(image)
        bboxes[:num_objects, [1, 3]] = 1 - bboxes[:num_objects, [3, 1]]

    # HORIZONTAL FLIP WITH PROBABILITY OF 0.5
    # SETTING PROBABILITY OF THE HORIZONTAL FLIP EXTERNALLY, BECAUSE OF BOXES!!!!!!
    if np.random.random() < prob_vertical_flip:
        transform2 = transforms.Compose([
                        # first ToPILImage and then ToTensor before normalize!!!!!!!!!!!!!!!!!!!!!!!!
                        transforms.ToPILImage(),
                        transforms.RandomHorizontalFlip(p = 1),
                        transforms.ToTensor()
                    ])
        image = transform2(image)
        bboxes[:num_objects, [0, 2]] = 1 - bboxes[:num_objects, [2, 0]]

    # GRAYSCALE WITH PROBABILITY 0.1, NO NEED TO CHANGE BOXES COORDINATES IN THIS CASE!!!!!!!!!!!!!!!!!!!!!!
    transform3 = transforms.Compose([
                    # first ToPILImage and then ToTensor before normalize!!!!!!!!!!!!!!!!!!!!!!!!
                    transforms.ToPILImage(),
                    transforms.RandomGrayscale(p = 0.1),
                    transforms.ToTensor()
                ])
    image = transform3(image)

    return image, bboxes
```

```python
# Starting with the architecture of the model. I will create one class for all 6 feature maps!!!!!

class AllFeatureMapsArchitecture(nn.Module):
    """
    From inputs to all six feature maps + downloading pretrained weights for the VGG part!!
    """
    def __init__(self):
        super(AllFeatureMapsArchitecture, self).__init__()

        # Standard convolutional layers in VGG16: first five convolutions

        # 3x300x300
        self.conv1_1 = nn.Conv2d(3, 64, kernel_size = 3, padding = 1)
        self.conv1_2 = nn.Conv2d(64, 64, kernel_size = 3, padding = 1)
        self.pool1 = nn.MaxPool2d(kernel_size = 2, stride = 2)

        # 64x150x150
        self.conv2_1 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.conv2_2 = nn.Conv2d(128, 128, kernel_size=3, padding=1)
        self.pool2 = nn.MaxPool2d(kernel_size=2, stride=2)

        # 128x75x75
        self.conv3_1 = nn.Conv2d(128, 256, kernel_size=3, padding=1)
        self.conv3_2 = nn.Conv2d(256, 256, kernel_size=3, padding=1)
        self.conv3_3 = nn.Conv2d(256, 256, kernel_size=3, padding=1)
        # to make from 75 to 38, not 37
        self.pool3 = nn.MaxPool2d(kernel_size=2, stride=2, ceil_mode=True)

        # 256x38x38
        self.conv4_1 = nn.Conv2d(256, 512, kernel_size=3, padding=1)
        self.conv4_2 = nn.Conv2d(512, 512, kernel_size=3, padding=1)
        self.conv4_3 = nn.Conv2d(512, 512, kernel_size=3, padding=1)
        self.pool4 = nn.MaxPool2d(kernel_size=2, stride=2)

        # 512x19x19
        self.conv5_1 = nn.Conv2d(512, 512, kernel_size=3, padding=1)
        self.conv5_2 = nn.Conv2d(512, 512, kernel_size=3, padding=1)
        self.conv5_3 = nn.Conv2d(512, 512, kernel_size=3, padding=1)
        # dimension of feature map stays the same
        self.pool5 = nn.MaxPool2d(kernel_size=3, stride=1, padding=1)


        # 2 layers which are substitutes to VGG16 fcs

        # atrous convolution  # 512x19x19
        self.conv6 = nn.Conv2d(512, 1024, kernel_size=3, padding=6, dilation=6)  
        self.conv7 = nn.Conv2d(1024, 1024, kernel_size=1)


        # Additional layers to generate 4 more feature maps (additional boxes) for prediction

        # 1024x19x19
        self.conv8_1 = nn.Conv2d(1024, 256, kernel_size = 1, padding = 0)
        self.conv8_2 = nn.Conv2d(256, 512, kernel_size = 3, stride = 2, padding = 1)

        # 512x10x10
        self.conv9_1 = nn.Conv2d(512, 128, kernel_size=1, padding=0)
        self.conv9_2 = nn.Conv2d(128, 256, kernel_size=3, stride=2, padding=1)

        # 256x5x5
        self.conv10_1 = nn.Conv2d(256, 128, kernel_size=1, padding=0)
        self.conv10_2 = nn.Conv2d(128, 256, kernel_size=3, padding=0)

        # 256x3x3
        self.conv11_1 = nn.Conv2d(256, 128, kernel_size=1, padding=0)
        self.conv11_2 = nn.Conv2d(128, 256, kernel_size=3, padding=0)


        # Loading pretrained weights for first 5 layers (VGG16)
        self.load_pretrained_layers()



    def load_pretrained_layers(self):
        """
        As in the paper, we use a VGG-16 pretrained on the ImageNet task as the base network.
        There's one available in PyTorch, see https://pytorch.org/docs/stable/torchvision/models.html#torchvision.models.vgg16
        We copy these parameters into our network. It's straightforward for conv1 to conv5.
        Other parameters are left as default values!!!
        """
        # Current state of base
        state_dict = self.state_dict()
        param_names = list(state_dict.keys())

        # Pretrained VGG base
        pretrained_state_dict = torchvision.models.vgg16(pretrained=True).state_dict()
        pretrained_param_names = list(pretrained_state_dict.keys())

        # Transfer conv. parameters from pretrained model to current model
        for i, param in enumerate(param_names[:26]):  # excluding conv6 and conv7 parameters
            state_dict[param] = pretrained_state_dict[pretrained_param_names[i]]

        self.load_state_dict(state_dict)

        print("\nLoaded base model.\n")


    ### IN THE LINK THERE IS ALSO THE INITIALIZATION OF THE PARAMETERS, BUT I SUPPOSE THAT THE DEFAULT ONE IS FINE, 
        ### NO NEED TO USE INIT_CONV2D FUNCTION!!!!!


    def forward(self, image):
        """
        Forward propagation.
        :param image: images, a tensor of dimensions (N, 3, 300, 300), N - batch size
        :return: 6 lower-level feature maps
        """
        conv1_out = self.pool1(F.relu(self.conv1_2(F.relu(self.conv1_1(image)))))
        conv2_out = self.pool2(F.relu(self.conv2_2(F.relu(self.conv2_1(conv1_out)))))
        conv3_out = self.pool3(F.relu(self.conv3_3(F.relu(self.conv3_2(F.relu(self.conv3_1(conv2_out)))))))
        first_feat_map = F.relu(self.conv4_3(F.relu(self.conv4_2(F.relu(self.conv4_1(conv3_out))))))
        conv4_out = self.pool4(first_feat_map)
        conv5_out = self.pool5(F.relu(self.conv5_3(F.relu(self.conv5_2(F.relu(self.conv5_1(conv4_out)))))))
        conv6_out = F.relu(self.conv6(conv5_out))
        conv7_out = F.relu(self.conv7(conv6_out))
        second_feat_map = conv7_out
        conv8_out = F.relu(self.conv8_2(F.relu(self.conv8_1(second_feat_map))))
        third_feat_map = conv8_out
        conv9_out = F.relu(self.conv9_2(F.relu(self.conv9_1(third_feat_map))))
        fourth_feat_map = conv9_out
        conv10_out = F.relu(self.conv10_2(F.relu(self.conv10_1(fourth_feat_map))))
        fifth_feat_map = conv10_out
        conv11_out = F.relu(self.conv11_2(F.relu(self.conv11_1(fifth_feat_map))))
        sixth_feat_map = conv11_out

        # returning all feature maps!!!!!!!
        return first_feat_map, second_feat_map, third_feat_map, fourth_feat_map, fifth_feat_map, sixth_feat_map
```

```python
# creating class which generates prediction vectors for all 6 feature maps!!!!

class PredictionConvolutions(nn.Module):
    """
    Convolutions to predict class scores and bounding boxes using lower and higher-level feature maps.
    The bounding boxes (locations) are predicted as encoded offsets w.r.t each of the 8732 prior (default) boxes.
    See 'cxcy_to_gcxgcy' in utils.py for the encoding definition.
    The class scores represent the scores of each object class in each of the 8732 bounding boxes located.
    A high score for 'background' = no object.
    """

    def __init__(self, n_classes):
        """
        :param n_classes: number of different types of objects. 4 in our case!!!
        """
        super(PredictionConvolutions, self).__init__()

        self.n_classes = n_classes

        # Number of prior-boxes we are considering per position in each feature map
        n_boxes = {'first_feat_map': 4,
                   'second_feat_map': 6,
                   'third_feat_map': 6,
                   'fourth_feat_map': 6,
                   'fifth_feat_map': 4,
                   'sixth_feat_map': 4}
        # 4 prior-boxes implies we use 4 different aspect ratios, etc.

        # Localization prediction convolutions (predict offsets w.r.t prior-boxes). 4 characteristics of location!!!!
        self.loc_first_feat_map = nn.Conv2d(512, n_boxes['first_feat_map'] * 4, kernel_size=3, padding=1)
        self.loc_second_feat_map = nn.Conv2d(1024, n_boxes['second_feat_map'] * 4, kernel_size=3, padding=1)
        self.loc_third_feat_map = nn.Conv2d(512, n_boxes['third_feat_map'] * 4, kernel_size=3, padding=1)
        self.loc_fourth_feat_map = nn.Conv2d(256, n_boxes['fourth_feat_map'] * 4, kernel_size=3, padding=1)
        self.loc_fifth_feat_map = nn.Conv2d(256, n_boxes['fifth_feat_map'] * 4, kernel_size=3, padding=1)
        self.loc_sixth_feat_map = nn.Conv2d(256, n_boxes['sixth_feat_map'] * 4, kernel_size=3, padding=1)

        # Class prediction convolutions (predict classes in localization boxes), n_classes = 4, including background
        self.cl_first_feat_map = nn.Conv2d(512, n_boxes['first_feat_map'] * n_classes, kernel_size=3, padding=1)
        self.cl_second_feat_map = nn.Conv2d(1024, n_boxes['second_feat_map'] * n_classes, kernel_size=3, padding=1)
        self.cl_third_feat_map = nn.Conv2d(512, n_boxes['third_feat_map'] * n_classes, kernel_size=3, padding=1)
        self.cl_fourth_feat_map = nn.Conv2d(256, n_boxes['fourth_feat_map'] * n_classes, kernel_size=3, padding=1)
        self.cl_fifth_feat_map = nn.Conv2d(256, n_boxes['fifth_feat_map'] * n_classes, kernel_size=3, padding=1)
        self.cl_sixth_feat_map = nn.Conv2d(256, n_boxes['sixth_feat_map'] * n_classes, kernel_size=3, padding=1)



    def forward(self, first_feat_map, second_feat_map, third_feat_map, fourth_feat_map, fifth_feat_map, sixth_feat_map):
        """
        Forward propagation.
        :param first_feat_map: conv4_3 feature map, a tensor of dimensions (N, 512, 38, 38)
        :param second_feat_map: conv7 feature map, a tensor of dimensions (N, 1024, 19, 19)
        :param third_feat_map: conv8_2 feature map, a tensor of dimensions (N, 512, 10, 10)
        :param fourth_feat_map: conv9_2 feature map, a tensor of dimensions (N, 256, 5, 5)
        :param fifth_feat_map: conv10_2 feature map, a tensor of dimensions (N, 256, 3, 3)
        :param sixth_feat_map: conv11_2 feature map, a tensor of dimensions (N, 256, 1, 1)
        :return: 8732 locations and class scores (i.e. w.r.t each prior box) for each image
        """
        batch_size = first_feat_map.size(0)

        # Predict localization boxes' bounds (as offsets w.r.t prior-boxes)
        l_first_feat_map = self.loc_first_feat_map(first_feat_map)  # (N, 16, 38, 38)
        l_first_feat_map = l_first_feat_map.permute(0, 2, 3,
                                      1).contiguous()  # (N, 38, 38, 16), to match prior-box order (after .view())
        # (.contiguous() ensures it is stored in a contiguous chunk of memory, needed for .view() below)!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
        l_first_feat_map = l_first_feat_map.view(batch_size, -1, 4)  # (N, 5776, 4), there are a total 5776 boxes on this feature map

        l_second_feat_map = self.loc_second_feat_map(second_feat_map)  # (N, 24, 19, 19)
        l_second_feat_map = l_second_feat_map.permute(0, 2, 3, 1).contiguous()  # (N, 19, 19, 24)
        l_second_feat_map = l_second_feat_map.view(batch_size, -1, 4)  # (N, 2166, 4), there are a total 2116 boxes on this feature map

        l_third_feat_map = self.loc_third_feat_map(third_feat_map)  # (N, 24, 10, 10)
        l_third_feat_map = l_third_feat_map.permute(0, 2, 3, 1).contiguous()  # (N, 10, 10, 24)
        l_third_feat_map = l_third_feat_map.view(batch_size, -1, 4)  # (N, 600, 4)

        l_fourth_feat_map = self.loc_fourth_feat_map(fourth_feat_map)  # (N, 24, 5, 5)
        l_fourth_feat_map = l_fourth_feat_map.permute(0, 2, 3, 1).contiguous()  # (N, 5, 5, 24)
        l_fourth_feat_map = l_fourth_feat_map.view(batch_size, -1, 4)  # (N, 150, 4)

        l_fifth_feat_map = self.loc_fifth_feat_map(fifth_feat_map)  # (N, 16, 3, 3)
        l_fifth_feat_map = l_fifth_feat_map.permute(0, 2, 3, 1).contiguous()  # (N, 3, 3, 16)
        l_fifth_feat_map = l_fifth_feat_map.view(batch_size, -1, 4)  # (N, 36, 4)

        l_sixth_feat_map = self.loc_sixth_feat_map(sixth_feat_map)  # (N, 16, 1, 1)
        l_sixth_feat_map = l_sixth_feat_map.permute(0, 2, 3, 1).contiguous()  # (N, 1, 1, 16)
        l_sixth_feat_map = l_sixth_feat_map.view(batch_size, -1, 4)  # (N, 4, 4)

        # Predict classes in localization boxes
        c_first_feat_map = self.cl_first_feat_map(first_feat_map)  # (N, 4 * n_classes, 38, 38)
        c_first_feat_map = c_first_feat_map.permute(0, 2, 3,
                                      1).contiguous()  # (N, 38, 38, 4 * n_classes), to match prior-box order (after .view())
        c_first_feat_map = c_first_feat_map.view(batch_size, -1,
                                   self.n_classes)  # (N, 5776, n_classes), there are a total 5776 boxes on this feature map

        c_second_feat_map = self.cl_second_feat_map(second_feat_map)  # (N, 6 * n_classes, 19, 19)
        c_second_feat_map = c_second_feat_map.permute(0, 2, 3, 1).contiguous()  # (N, 19, 19, 6 * n_classes)
        c_second_feat_map = c_second_feat_map.view(batch_size, -1,
                               self.n_classes)  # (N, 2166, n_classes), there are a total 2116 boxes on this feature map

        c_third_feat_map = self.cl_third_feat_map(third_feat_map)  # (N, 6 * n_classes, 10, 10)
        c_third_feat_map = c_third_feat_map.permute(0, 2, 3, 1).contiguous()  # (N, 10, 10, 6 * n_classes)
        c_third_feat_map = c_third_feat_map.view(batch_size, -1, self.n_classes)  # (N, 600, n_classes)

        c_fourth_feat_map = self.cl_fourth_feat_map(fourth_feat_map)  # (N, 6 * n_classes, 5, 5)
        c_fourth_feat_map = c_fourth_feat_map.permute(0, 2, 3, 1).contiguous()  # (N, 5, 5, 6 * n_classes)
        c_fourth_feat_map = c_fourth_feat_map.view(batch_size, -1, self.n_classes)  # (N, 150, n_classes)

        c_fifth_feat_map = self.cl_fifth_feat_map(fifth_feat_map)  # (N, 4 * n_classes, 3, 3)
        c_fifth_feat_map = c_fifth_feat_map.permute(0, 2, 3, 1).contiguous()  # (N, 3, 3, 4 * n_classes)
        c_fifth_feat_map = c_fifth_feat_map.view(batch_size, -1, self.n_classes)  # (N, 36, n_classes)

        c_sixth_feat_map = self.cl_sixth_feat_map(sixth_feat_map)  # (N, 4 * n_classes, 1, 1)
        c_sixth_feat_map = c_sixth_feat_map.permute(0, 2, 3, 1).contiguous()  # (N, 1, 1, 4 * n_classes)
        c_sixth_feat_map = c_sixth_feat_map.view(batch_size, -1, self.n_classes)  # (N, 4, n_classes)

        # A total of 8732 boxes
        # Concatenate in this specific order (i.e. must match the order of the prior-boxes)
        locs = torch.cat([l_first_feat_map, l_second_feat_map, l_third_feat_map, l_fourth_feat_map, l_fifth_feat_map, l_sixth_feat_map], dim=1)  # (N, 8732, 4)
        classes_scores = torch.cat([c_first_feat_map, c_second_feat_map, c_third_feat_map, c_fourth_feat_map, c_fifth_feat_map, c_sixth_feat_map], dim=1)  # (N, 8732, n_classes)

        return locs, classes_scores
```

```python
# some functions which are needed for the classes below!!!!!!
# transforming coordinates of the boxes (and vice-versa counterparts) !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
# xy - original, cxcy - transformed (centered), gcxgcy - normalized which are returned in predicion!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
def xy_to_cxcy(xy):
    """
    Convert bounding boxes from boundary coordinates (x_min, y_min, x_max, y_max) to center-size coordinates (c_x, c_y, w, h).
    :param xy: bounding boxes in boundary coordinates, a tensor of size (n_boxes, 4)
    :return: bounding boxes in center-size coordinates, a tensor of size (n_boxes, 4)
    """
    return torch.cat([(xy[:, 2:] + xy[:, :2]) / 2,  # c_x, c_y
                      xy[:, 2:] - xy[:, :2]], 1)  # w, h


def cxcy_to_xy(cxcy):
    """
    Convert bounding boxes from center-size coordinates (c_x, c_y, w, h) to boundary coordinates (x_min, y_min, x_max, y_max).
    :param cxcy: bounding boxes in center-size coordinates, a tensor of size (n_boxes, 4)
    :return: bounding boxes in boundary coordinates, a tensor of size (n_boxes, 4)
    """
    return torch.cat([cxcy[:, :2] - (cxcy[:, 2:] / 2),  # x_min, y_min
                      cxcy[:, :2] + (cxcy[:, 2:] / 2)], 1)  # x_max, y_max


def cxcy_to_gcxgcy(cxcy, priors_cxcy):
    """
    Encode bounding boxes (that are in center-size form) w.r.t. the corresponding prior boxes (that are in center-size form).
    For the center coordinates, find the offset with respect to the prior box, and scale by the size of the prior box.
    For the size coordinates, scale by the size of the prior box, and convert to the log-space.
    In the model, we are predicting bounding box coordinates in this encoded form.
    :param cxcy: bounding boxes in center-size coordinates, a tensor of size (n_priors, 4)
    :param priors_cxcy: prior boxes with respect to which the encoding must be performed, a tensor of size (n_priors, 4)
    :return: encoded bounding boxes, a tensor of size (n_priors, 4)
    """

    # The 10 and 5 below are referred to as 'variances' in the original Caffe repo, completely empirical
    # They are for some sort of numerical conditioning, for 'scaling the localization gradient'
    # See https://github.com/weiliu89/caffe/issues/155
    return torch.cat([(cxcy[:, :2] - priors_cxcy[:, :2]) / (priors_cxcy[:, 2:] / 10),  # g_c_x, g_c_y
                      torch.log((cxcy[:, 2:]) / priors_cxcy[:, 2:]) * 5], 1)  # g_w, g_h


def gcxgcy_to_cxcy(gcxgcy, priors_cxcy):
    """
    Decode bounding box coordinates predicted by the model, since they are encoded in the form mentioned above.
    They are decoded into center-size coordinates.
    This is the inverse of the function above.
    :param gcxgcy: encoded bounding boxes, i.e. output of the model, a tensor of size (n_priors, 4)
    :param priors_cxcy: prior boxes with respect to which the encoding is defined, a tensor of size (n_priors, 4)
    :return: decoded bounding boxes in center-size form, a tensor of size (n_priors, 4)
    """

    return torch.cat([gcxgcy[:, :2] * priors_cxcy[:, 2:] / 10 + priors_cxcy[:, :2],  # c_x, c_y
                      torch.exp(gcxgcy[:, 2:] / 5) * priors_cxcy[:, 2:]], 1)  # w, h



# Functions which are helpful to calculate Jaccard Index between each pair of the boxes given (Kind of Jaccard Index matrix)!!!!
def find_intersection(set_1, set_2):
    """
    Find the intersection of every box combination between two sets of boxes that are in boundary coordinates.
    :param set_1: set 1, a tensor of dimensions (n1, 4)
    :param set_2: set 2, a tensor of dimensions (n2, 4)
    :return: intersection of each of the boxes in set 1 with respect to each of the boxes in set 2, a tensor of dimensions (n1, n2)
    """

    # PyTorch auto-broadcasts singleton dimensions
    lower_bounds = torch.max(set_1[:, :2].unsqueeze(1), set_2[:, :2].unsqueeze(0))  # (n1, n2, 2)
    upper_bounds = torch.min(set_1[:, 2:].unsqueeze(1), set_2[:, 2:].unsqueeze(0))  # (n1, n2, 2)
    intersection_dims = torch.clamp(upper_bounds - lower_bounds, min=0)  # (n1, n2, 2)
    return intersection_dims[:, :, 0] * intersection_dims[:, :, 1]  # (n1, n2)


def find_jaccard_overlap(set_1, set_2):
    """
    Find the Jaccard Overlap (IoU) of every box combination between two sets of boxes that are in boundary coordinates.
    :param set_1: set 1, a tensor of dimensions (n1, 4)
    :param set_2: set 2, a tensor of dimensions (n2, 4)
    :return: Jaccard Overlap of each of the boxes in set 1 with respect to each of the boxes in set 2, a tensor of dimensions (n1, n2)
    """

    # Find intersections
    intersection = find_intersection(set_1, set_2)  # (n1, n2)

    # Find areas of each box in both sets
    areas_set_1 = (set_1[:, 2] - set_1[:, 0]) * (set_1[:, 3] - set_1[:, 1])  # (n1)
    areas_set_2 = (set_2[:, 2] - set_2[:, 0]) * (set_2[:, 3] - set_2[:, 1])  # (n2)

    # Find the union
    # PyTorch auto-broadcasts singleton dimensions
    union = areas_set_1.unsqueeze(1) + areas_set_2.unsqueeze(0) - intersection  # (n1, n2)

    return intersection / union  # (n1, n2)
```

```python
# MODEL CLASS!!!!!
# Joining Feature maps creation and prediction convolution + creation of initial default 8732 boxes (prior boxes)!!! +
    # + all sorted boxes return, which predict some class with high-enough score and are not deleted after NMS + top scores only left!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!

class SSD300(nn.Module):
    """
    The SSD300 network - encapsulates all feature maps and prediction convolutions.
    """

    def __init__(self, n_classes):
        super(SSD300, self).__init__()

        self.n_classes = n_classes

        self.feat_maps = AllFeatureMapsArchitecture()
        self.pred_convs = PredictionConvolutions(n_classes)

        # Since lower level features (first_feat_map) have considerably larger scales, we take the L2 norm and rescale!!!!!!!!!!!!!!!!!!!!
        # Rescale factor is initially set at 20, but is learned for each channel during back-prop!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
        self.rescale_factors = nn.Parameter(torch.FloatTensor(1, 512, 1, 1))  # there are 512 channels in first_feat_map
        nn.init.constant_(self.rescale_factors, 20)

        # Prior boxes
        self.priors_cxcy = self.create_prior_boxes()


    def forward(self, image):
        """
        Forward propagation.
        :param image: images, a tensor of dimensions (N, 3, 300, 300)
        :return: 8732 locations and class scores (i.e. w.r.t each prior box) for each image
        """
        # All 6 feature maps creation
        # (N, 512, 38, 38), (N, 1024, 19, 19), (N, 512, 10, 10), (N, 256, 5, 5), (N, 256, 3, 3), (N, 256, 1, 1)
        first_feat_map, second_feat_map, third_feat_map, fourth_feat_map, fifth_feat_map, sixth_feat_map = self.feat_maps(image)  

        # Rescale first_feat_map after L2 norm
        norm = first_feat_map.pow(2).sum(dim=1, keepdim=True).sqrt()  # (N, 1, 38, 38)
        first_feat_map = first_feat_map / norm  # (N, 512, 38, 38)
        first_feat_map = first_feat_map * self.rescale_factors  # (N, 512, 38, 38)
        # (PyTorch autobroadcasts singleton dimensions during arithmetic)

        # Run prediction convolutions (predict offsets w.r.t prior-boxes and classes in each resulting localization box)
        locs, classes_scores = self.pred_convs(first_feat_map, second_feat_map, third_feat_map, fourth_feat_map, fifth_feat_map, 
                                               sixth_feat_map)  # (N, 8732, 4), (N, 8732, n_classes)

        return locs, classes_scores


    def create_prior_boxes(self):
        """
        Create the 8732 prior (default) boxes for the SSD300, as defined in the paper.
        :return: prior boxes in center-size coordinates, a tensor of dimensions (8732, 4)
        """
        fmap_dims = {'first_feat_map': 38,
                     'second_feat_map': 19,
                     'third_feat_map': 10,
                     'fourth_feat_map': 5,
                     'fifth_feat_map': 3,
                     'sixth_feat_map': 1}

        obj_scales = {'first_feat_map': 0.1,
                      'second_feat_map': 0.2,
                      'third_feat_map': 0.375,
                      'fourth_feat_map': 0.55,
                      'fifth_feat_map': 0.725,
                      'sixth_feat_map': 0.9}

        aspect_ratios = {'first_feat_map': [1., 2., 0.5],
                         'second_feat_map': [1., 2., 3., 0.5, .333],
                         'third_feat_map': [1., 2., 3., 0.5, .333],
                         'fourth_feat_map': [1., 2., 3., 0.5, .333],
                         'fifth_feat_map': [1., 2., 0.5],
                         'sixth_feat_map': [1., 2., 0.5]}

        fmaps = list(fmap_dims.keys())

        prior_boxes = []

        for k, fmap in enumerate(fmaps):
            for i in range(fmap_dims[fmap]):
                for j in range(fmap_dims[fmap]):
                    # center of the box
                    cx = (j + 0.5) / fmap_dims[fmap]
                    cy = (i + 0.5) / fmap_dims[fmap]

                    for ratio in aspect_ratios[fmap]:
                        # adding the center coordinates, width and height for each box!!!!!
                        prior_boxes.append([cx, cy, obj_scales[fmap] * sqrt(ratio), obj_scales[fmap] / sqrt(ratio)])

                        # For an aspect ratio of 1, use an additional prior whose scale is the geometric mean of the
                        # scale of the current feature map and the scale of the next feature map
                        if ratio == 1.:
                            try:
                                additional_scale = sqrt(obj_scales[fmap] * obj_scales[fmaps[k + 1]])
                            # For the last feature map, there is no "next" feature map
                            except IndexError:
                                additional_scale = 1.
                            prior_boxes.append([cx, cy, additional_scale, additional_scale])

        prior_boxes = torch.FloatTensor(prior_boxes).to(device)  # (8732, 4)
        # all corners of the box should be inside the photo!!!!!!
        prior_boxes.clamp_(0, 1)  # (8732, 4)

        return prior_boxes


    def detect_objects(self, predicted_locs, predicted_scores, min_score, max_overlap, top_k):
        """
        Decipher the 8732 locations and class scores (output of ths SSD300) to detect objects.
        For each class, perform Non-Maximum Suppression (NMS) on boxes that are above a minimum threshold.
        :param predicted_locs: predicted locations/boxes w.r.t the 8732 prior boxes, a tensor of dimensions (N, 8732, 4)
        :param predicted_scores: class scores for each of the encoded locations/boxes, a tensor of dimensions (N, 8732, n_classes)
        :param min_score: minimum threshold for a box to be considered a match for a certain class
        :param max_overlap: maximum overlap two boxes can have so that the one with the lower score is not suppressed via NMS
        :param top_k: if there are a lot of resulting detection across all classes, keep only the top 'k'
        :return: detections (boxes, labels, and scores), lists of length batch_size
        """
        batch_size = predicted_locs.size(0)
        n_priors = self.priors_cxcy.size(0)
        predicted_scores = F.softmax(predicted_scores, dim=2)  # (N, 8732, n_classes)

        # Lists to store final predicted boxes, labels, and scores for all images
        all_images_boxes = list()
        all_images_labels = list()
        all_images_scores = list()

        assert n_priors == predicted_locs.size(1) == predicted_scores.size(1)

        for i in range(batch_size):
            # Decode object coordinates from the form we regressed predicted boxes to
            decoded_locs = cxcy_to_xy(
                gcxgcy_to_cxcy(predicted_locs[i], self.priors_cxcy))  # (8732, 4), these are fractional pt. coordinates

            # Lists to store boxes and scores for this image
            image_boxes = list()
            image_labels = list()
            image_scores = list()

            max_scores, best_label = predicted_scores[i].max(dim=1)  # (8732)

            # Check for each class
            for c in range(1, self.n_classes):
                # Keep only predicted boxes and scores where scores for this class are above the minimum score
                class_scores = predicted_scores[i][:, c]  # (8732)
                score_above_min_score = class_scores > min_score  # torch.uint8 (byte) tensor, for indexing
                n_above_min_score = score_above_min_score.sum().item()
                if n_above_min_score == 0:
                    continue
                class_scores = class_scores[score_above_min_score]  # (n_qualified), n_min_score <= 8732
                class_decoded_locs = decoded_locs[score_above_min_score]  # (n_qualified, 4)

                # Sort predicted boxes and scores by scores
                class_scores, sort_ind = class_scores.sort(dim=0, descending=True)  # (n_qualified), (n_min_score)
                class_decoded_locs = class_decoded_locs[sort_ind]  # (n_min_score, 4)

                # Find the overlap between predicted boxes
                overlap = find_jaccard_overlap(class_decoded_locs, class_decoded_locs)  # (n_qualified, n_min_score)

                # Non-Maximum Suppression (NMS)

                # A torch.uint8 (byte) tensor to keep track of which predicted boxes to suppress
                # 1 implies suppress, 0 implies don't suppress
                suppress = torch.zeros((n_above_min_score), dtype=torch.uint8).to(device)  # (n_qualified)

                # Consider each box in order of decreasing scores
                for box in range(class_decoded_locs.size(0)):
                    # If this box is already marked for suppression
                    if suppress[box] == 1:
                        continue

                    # Suppress boxes whose overlaps (with this box) are greater than maximum overlap
                    # Find such boxes and update suppress indices
                    suppress = torch.max(suppress, overlap[box] > max_overlap)
                    # The max operation retains previously suppressed boxes, like an 'OR' operation

                    # Don't suppress this box, even though it has an overlap of 1 with itself
                    suppress[box] = 0

                # Store only unsuppressed boxes for this class
                image_boxes.append(class_decoded_locs[1 - suppress])
                image_labels.append(torch.LongTensor((1 - suppress).sum().item() * [c]).to(device))
                image_scores.append(class_scores[1 - suppress])

            # If no object in any class is found, store a placeholder for 'background'
            if len(image_boxes) == 0:
                image_boxes.append(torch.FloatTensor([[0., 0., 1., 1.]]).to(device))
                image_labels.append(torch.LongTensor([0]).to(device))
                image_scores.append(torch.FloatTensor([0.]).to(device))

            # Concatenate into single tensors
            image_boxes = torch.cat(image_boxes, dim=0)  # (n_objects, 4)
            image_labels = torch.cat(image_labels, dim=0)  # (n_objects)
            image_scores = torch.cat(image_scores, dim=0)  # (n_objects)
            n_objects = image_scores.size(0)

            # Keep only the top k objects
            if n_objects > top_k:
                image_scores, sort_ind = image_scores.sort(dim=0, descending=True)
                image_scores = image_scores[:top_k]  # (top_k)
                image_boxes = image_boxes[sort_ind][:top_k]  # (top_k, 4)
                image_labels = image_labels[sort_ind][:top_k]  # (top_k)

            # Append to lists that store predicted boxes and scores for all images
            all_images_boxes.append(image_boxes)
            all_images_labels.append(image_labels)
            all_images_scores.append(image_scores)

        return all_images_boxes, all_images_labels, all_images_scores  # lists of length batch_size
```

```python
# Class with calculates the mains loss function which is used in training of the model!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!

class MultiBoxLoss(nn.Module):
    """
    The MultiBox loss, a loss function for object detection.
    This is a combination of:
    (1) a localization loss for the predicted locations of the boxes, and
    (2) a confidence loss for the predicted class scores.
    """

    def __init__(self, priors_cxcy, threshold=0.5, neg_pos_ratio=3, alpha=1.):
        super(MultiBoxLoss, self).__init__()
        self.priors_cxcy = priors_cxcy
        self.priors_xy = cxcy_to_xy(priors_cxcy)
        self.threshold = threshold
        self.neg_pos_ratio = neg_pos_ratio
        self.alpha = alpha

        self.smooth_l1 = nn.L1Loss()
        self.cross_entropy = nn.CrossEntropyLoss(reduce=False)

    def forward(self, predicted_locs, predicted_scores, boxes, labels):
        """
        Forward propagation.
        :param predicted_locs: predicted locations/boxes w.r.t the 8732 prior boxes, a tensor of dimensions (N, 8732, 4)
        :param predicted_scores: class scores for each of the encoded locations/boxes, a tensor of dimensions (N, 8732, n_classes)
        :param boxes: true object bounding boxes in boundary coordinates, a list of N tensors
        :param labels: true object labels, a list of N tensors
        :return: multibox loss, a scalar
        """
        batch_size = predicted_locs.size(0)
        n_priors = self.priors_cxcy.size(0)
        n_classes = predicted_scores.size(2)

        assert n_priors == predicted_locs.size(1) == predicted_scores.size(1)

        true_locs = torch.zeros((batch_size, n_priors, 4), dtype=torch.float).to(device)  # (N, 8732, 4)
        true_classes = torch.zeros((batch_size, n_priors), dtype=torch.long).to(device)  # (N, 8732)

        # For each image
        for i in range(batch_size):

            # getting rid of redundant boxes and labels, which were artificially added to make the batching process work!!!!!!!!!!!!!!!!!!!
            list_filter = []
            for j in range(len(boxes[i])):
                if boxes[i][j].max() != 0:
                    list_filter.append(j)
                else:
                    break
            filt_boxes = boxes[i][list_filter]
            filt_labels = labels[i][list_filter]


            n_objects = filt_boxes.size(0)

            # DON'T FORGET TO TAKE ONLY NON-ZERO BOXES AND THE CORRESPONDING LABELS (THERE ARE REDUNDANT BOXES TO MAKE EACH SAMPLE EQUAL SHAPE, 
                # TO MAKE SAMPLES BATCHED)!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
            overlap = find_jaccard_overlap(filt_boxes,
                                           self.priors_xy)  # (n_objects, 8732)

            # For each prior, find the object that has the maximum overlap!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
            overlap_for_each_prior, object_for_each_prior = overlap.max(dim=0)  # (8732)

            # We don't want a situation where an object is not represented in our positive (non-background) priors -
            # 1. An object might not be the best object for all priors, and is therefore not in object_for_each_prior.
            # 2. All priors with the object may be assigned as background based on the threshold (0.5).

            # To remedy this -
            # First, find the prior that has the maximum overlap for each object. (Remedy of 1.)!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
            _, prior_for_each_object = overlap.max(dim=1)  # (N_objects)

            # Then, assign each object to the corresponding maximum-overlap-prior. (This fixes 1.)
            # THIS IS SO BECAUSE IN PRIOR_FOR_EACH_OBJECT, FIRST ELEMENT CORRESPONDS TO FIRST CLASS AND SO ON 
                # (THAT'S HOW JACCARD INDEX IS CREATED)!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
            object_for_each_prior[prior_for_each_object] = torch.LongTensor(range(n_objects)).to(device)

            # To ensure these priors qualify, artificially give them an overlap of greater than 0.5. (This fixes 2.)!!!!!!!!!!!!!
            overlap_for_each_prior[prior_for_each_object] = 1.

            # Labels for each prior
            # LABELS[I] - ONE PHOTO - TENSOR OF LABELS FOR EACH OBJECT
            # OBJECT_FOR_EACH_PRIOR - ASSIGNING THE PRIOR TO THE OBJECT(BOX) != LABEL!!!!!! (INDEX OF THE LABEL IN LABELS[I])!!!!!!!!!!!!
            # LABELS[I][OBJECT_FOR_EACH_PRIOR] FINDS LABEL OF OBJECT BY INDEX OF THE LABEL IN LABELS[I] FOR EACH PRIOR
            label_for_each_prior = filt_labels[object_for_each_prior]  # (8732)
            # Set priors whose overlaps with objects are less than the threshold to be background (no object)!!!!!!!!!!!!!!!!!!!!!!!!!!!!
            # SO FOR THESE PRIOR BOXES THE LABEL IS 0, AND INDEX OF THE BACKGROUND IN THE GLOBAL VARIABLE 'CLASSES' IS 0 AS WELL!!!!!!!!!!!!
            label_for_each_prior[overlap_for_each_prior < self.threshold] = 0  # (8732)
            # Store
            true_classes[i] = label_for_each_prior[:, 0]
            # Encode center-size object coordinates into the form we regressed predicted boxes to!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
            true_locs[i] = cxcy_to_gcxgcy(xy_to_cxcy(boxes[i][object_for_each_prior]), self.priors_cxcy)  # (8732, 4)

        # Identify priors that are positive (object/non-background)
        positive_priors = true_classes != 0  # (N, 8732)

        # LOCALIZATION LOSS
        # Localization loss is computed only over positive (non-background) priors
        loc_loss = self.smooth_l1(predicted_locs[positive_priors], true_locs[positive_priors])  # (), scalar

        # Note: indexing with a torch.uint8 (byte) tensor flattens the tensor when indexing is across multiple dimensions (N & 8732)
        # So, if predicted_locs has the shape (N, 8732, 4), predicted_locs[positive_priors] will have (total positives, 4)

        # CONFIDENCE LOSS

        # Confidence loss is computed over positive priors and the most difficult (hardest) negative priors in each image
        # That is, FOR EACH IMAGE,
        # we will take the hardest (neg_pos_ratio * n_positives) negative priors, i.e where there is maximum loss
        # This is called Hard Negative Mining - it concentrates on hardest negatives in each image, and also minimizes pos/neg imbalance

        # Number of positive and hard-negative priors per image
        n_positives = positive_priors.sum(dim=1)  # (N)
        n_hard_negatives = self.neg_pos_ratio * n_positives  # (N)
        # First, find the loss for all priors
        # IT'S FINE THAT LABEL IS IN THE N*8732 FORMAT, IT SHOULD BE LIKE THAT, I.E NO NEED TO DO 2 -> [0, 0, 1, 0]!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
        conf_loss_all = self.cross_entropy(predicted_scores.view(-1, n_classes), true_classes.view(-1))  # (N * 8732)
        conf_loss_all = conf_loss_all.view(batch_size, n_priors)  # (N, 8732)

        # We already know which priors are positive
        conf_loss_pos = conf_loss_all[positive_priors]  # (sum(n_positives))

        # Next, find which priors are hard-negative
        # To do this, sort ONLY negative priors in each image in order of decreasing loss and take top n_hard_negatives
        conf_loss_neg = conf_loss_all.clone()  # (N, 8732)
        conf_loss_neg[positive_priors] = 0.  # (N, 8732), positive priors are ignored (never in top n_hard_negatives)
        conf_loss_neg, _ = conf_loss_neg.sort(dim=1, descending=True)  # (N, 8732), sorted by decreasing hardness
        # CREATING N INCREASING SEQUENCES !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
        hardness_ranks = torch.LongTensor(range(n_priors)).unsqueeze(0).expand_as(conf_loss_neg).to(device)  # (N, 8732)
        # BOOL, 1, IF RANK IS LESS THAN NUMBER OF HARD NEGATIVES REQUIRED!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
        hard_negatives = hardness_ranks < n_hard_negatives.unsqueeze(1)  # (N, 8732)
        # TAKING ONLY CERTAIN NUMBER OF HARD NEGATIVES, BUT WITH HIGHEST LOSSES (CONF_LOSS_NEG IS SORTED)!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
        conf_loss_hard_neg = conf_loss_neg[hard_negatives]  # (sum(n_hard_negatives))
        # As in the paper, averaged over positive priors only, although computed over both positive and hard-negative priors
        conf_loss = (conf_loss_hard_neg.sum() + conf_loss_pos.sum()) / n_positives.sum().float()  # (), scalar

        # TOTAL LOSS
        return conf_loss + self.alpha * loc_loss
```

```python
# TRAINING FUNCTION!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
# NOTE: I DIDN'T GET HOW TO IMPLEMENT TENSORBOARD IN COLLAB, SO I WILL JUST PRINT LOSS FUNCTIONS AND LOSS GRAPHS EVERY 5-10 EPOCHS!!!!!!!!!!!!!!!!!!!!!!!
# ANOTHER NOTE: I WANTED TO INCLUDE IN THE EPOCH OUTPUT SOME METRICS CALCULATION, BUT THE CALCULATION OF SUCH METRICS IS A LITTLE BIT 
    # TIME-CONSUMING, SO I'VE DECIDED TO CALCULATE THE METRICS FOR THE FINAL MODEL ONLY LATER ON!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!

def train(model, criterion, optimizer, scheduler, n_epochs): 

    train_losses = []
    # HERE TEST SET IS LIKE THE VALIDATION ONE!!!
    test_losses = []

    for epoch in tqdm_notebook(range(n_epochs)):
        model.train()
        train_losses_per_batch = []
        for batch in tqdm(train_loader):

            optimizer.zero_grad()

            images = batch['image'].to(device)
            boxes = [b.to(device) for b in batch['bboxes']]
            labels = [l.to(device) for l in batch['label']]
            num_objects_batch = batch['num_b']
            # AUGMENTATIONS!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
            # basically images.shape[0] is the batch_size!!!!
            for j in range(images.shape[0]):
                num_objects = num_objects_batch[j].item()
                # AUGMENTATIONS, SO THAT EACH PHOTO MAY BE AUGMENTED DIFFERENTLY IN DIFFERENT EPOCHS!!!!!!!!!!!!!!!!!!!!!!!!!!!!!            
                images[j, :, :, :], boxes[j] = general_augmentation(images[j, :, :, :], boxes[j], num_objects)
            
            # NORMALIZING PHOTOS, SO THAT MEAN 0 AND STD 1 PER CHANNEL!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
            # INITIATED AT THE BEGINNING OF THE PROJECT GLOBAL VARIABLES MEAN_PER_CHANNEL AND STD_PER_CHANNEL WERE USED!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
            transform1 = transforms.Compose([
                transforms.Normalize(mean_per_channel, std_per_channel) 
            ])
            images = transform1(images)

            predicted_locs, predicted_scores = model(images)
            loss = criterion(predicted_locs, predicted_scores, boxes, labels)
            loss.backward()
            optimizer.step()
            train_losses_per_batch.append(loss.item())

        scheduler.step()

        epoch_train_loss = np.round(np.mean(train_losses_per_batch), 3)
        train_losses.append(epoch_train_loss)

        # HERE TEST SET IS LIKE THE VALIDATION ONE!!!
        model.eval()
        test_losses_per_epoch = []
        with torch.no_grad():
            for batch in tqdm(test_loader):
                # NORMALIZING VALIDATION IMAGES AS WELL!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
                images = transform1(batch['image'].to(device))
                boxes = [b.to(device) for b in batch['bboxes']]
                labels = [l.to(device) for l in batch['label']]
                predicted_locs, predicted_scores = model(images)
                loss = criterion(predicted_locs, predicted_scores, boxes, labels)
                test_losses_per_epoch.append(loss.item())

        epoch_test_loss = np.round(np.mean(test_losses_per_epoch), 3)
        test_losses.append(epoch_test_loss)

        # PRINTING LOSS DATA EVERY EPOCH!!!!
        print('Train Loss epoch {}: {}'.format(epoch + 1, epoch_train_loss))
        print('Test Loss epoch {}: {}'.format(epoch + 1, epoch_test_loss))
        print('List of train losses after each epoch')
        print(train_losses)
        print('List of test losses after each epoch')
        print(test_losses)
        
        # EVERY TENTH EPOCH!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
        if (epoch + 1) % 10 == 0:
            # SAVING MODEL WEIGHTS TO THE DISK!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
            filename = '/content/drive/MyDrive/project_DLS/model_after_{}_epochs1.pth'.format(epoch + 1)
            torch.save(model.state_dict(), filename)

            # SHOWING LOSS GRAPHS!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
            plt.figure(figsize=(12, 6))
            plt.plot(train_losses, label="train_loss")
            plt.plot(test_losses, label="val_loss")
            plt.legend(loc='best', fontsize = 14)
            plt.xlabel("epochs", fontsize = 16)
            plt.ylabel("loss", fontsize = 16)
            #plt.ylim(1, 8)
            plt.title('Train and Validation losses dynamics after {} epochs'.format(epoch + 1), fontsize = 20)
            plt.show()


    return train_losses, test_losses
```

**6). Измерение качества работы модели (метрики согласуются с руководителем и зависят от задачи)**

Подробнее про них можно послушать в [видеолекции 2018 года](https://www.youtube.com/watch?v=ewkSI2cuyoQ&list=PL0Ks75aof3ThkitsZbUOEQg7Ybl5kB_s3&index=24).

Необходимо самостоятельно реализовать функции, вычисляющие поклассовые Precision и Recall. На вход функциям поступают веса модели и выборка, на которой нужно измерить метрики. На выходе $-$ таблица с precision и recall для каждого класса.

> Результат пункта $-$ реализованные функции метрик для задачи детектирования, позволяющие оценить качество работы модели на выборке. Выборка/даталоадер является аргументом функций.

#### CALCULATING 1) MAP, 2) PER CLASS JACCARD INDEX FOR TP (TRUE POSITIVES)!!! AND 3) PRECISION/RECALL/F1-SCORE PER CLASS!!!!

#### NOTE: THE INPUT IN THE FUNCTION BELOW, CONTAINS REDUNDANT LABELS AND BOXES, BUT THIS IS FINE, AS IN THE FUNCTION THERE IS LOOPING FOR ONLY 3 CLASSES ONLY, SO THESE REDUNDANT LABELS WILL NOT BE CONSIDERED AND RESULT WILL BE UNAFFECTED!!!!!!!

```python
# CALCULATING 1)MAP, 2)PER CLASS JACCARD INDEX FOR TP!!! AND 3)PRECISION/RECALL/F1-SCORE PER CLASS!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!

def calculate_mAP_PCavJTP_PCPrRe(det_boxes, det_labels, det_scores, true_boxes, true_labels):
    """
    Calculate the Mean Average Precision (mAP) of detected objects.
    See https://medium.com/@jonathan_hui/map-mean-average-precision-for-object-detection-45c121a31173 for an explanation
    :param det_boxes: list of tensors, one tensor for each image containing detected objects' bounding boxes
    :param det_labels: list of tensors, one tensor for each image containing detected objects' labels
    :param det_scores: list of tensors, one tensor for each image containing detected objects' labels' scores
    :param true_boxes: list of tensors, one tensor for each image containing actual objects' bounding boxes
    :param true_labels: list of tensors, one tensor for each image containing actual objects' labels
    :return: list of average precisions for all classes, mean average precision (mAP)
    """
    assert len(det_boxes) == len(det_labels) == len(det_scores) == len(true_boxes) == len(
        true_labels)  # these are all lists of tensors of the same length, i.e. number of images

    # CLASSES IS THE GLOBAL VARIABLE INITIALIZED AT THE BEGINNING!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    n_classes = len(classes)

    # Store all (true) objects in a single continuous tensor while keeping track of the image it is from
    true_images = list()
    for i in range(len(true_labels)):
        true_images.extend([i] * true_labels[i].size(0))
    true_images = torch.LongTensor(true_images).to(
        device)  # (n_objects), n_objects is the total no. of objects across all images
    true_boxes = torch.cat(true_boxes, dim=0)  # (n_objects, 4)
    true_labels = torch.cat(true_labels, dim=0).view(-1)  # (n_objects)

    assert true_images.size(0) == true_boxes.size(0) == true_labels.size(0)

    # Store all detections in a single continuous tensor while keeping track of the image it is from
    det_images = list()
    for i in range(len(det_labels)):
        det_images.extend([i] * det_labels[i].size(0))
    det_images = torch.LongTensor(det_images).to(device)  # (n_detections)
    det_boxes = torch.cat(det_boxes, dim=0)  # (n_detections, 4)
    det_labels = torch.cat(det_labels, dim=0)  # (n_detections)
    det_scores = torch.cat(det_scores, dim=0)  # (n_detections)

    assert det_images.size(0) == det_boxes.size(0) == det_labels.size(0) == det_scores.size(0)

    # Calculate APs for each class (except background)
    average_precisions = torch.zeros((n_classes - 1), dtype=torch.float)  # (n_classes - 1)
    # CALCULAITNG AVERAGE JACCARD AMONG TRUE POSITIVES FOR EACH CLASS
    jaccard_tp_average = torch.zeros((n_classes - 1), dtype=torch.float)  # (n_classes - 1)
    # PRECISION ,RECALL AND F1-SCORE PER CLASS (THRESHOLD 0.5)!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    simple_precisions = torch.zeros((n_classes - 1), dtype=torch.float).to(device)
    simple_recalls = torch.zeros((n_classes - 1), dtype=torch.float).to(device)
    simple_f1_scores = torch.zeros((n_classes - 1), dtype=torch.float).to(device)

    # STARTING FROM 1 AS BACKGROUND HAS LABEL 0!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    for c in range(1, n_classes):
        # Extract only objects with this class
        true_class_images = true_images[true_labels == c]  # (n_class_objects)
        true_class_boxes = true_boxes[true_labels == c]  # (n_class_objects, 4)
        n_class_objects = len(true_class_images) # SCALAR, NUMBER OF THE CLASS OBJECTS OVERALL!!!!

        # Keep track of which true objects with this class have already been 'detected'
        # So far, none
        true_class_boxes_detected = torch.zeros((n_class_objects), dtype=torch.uint8).to(device)  # (n_class_objects)

        # Extract only detections with this class
        det_class_images = det_images[det_labels == c]  # (n_class_detections)
        det_class_boxes = det_boxes[det_labels == c]  # (n_class_detections, 4)
        det_class_scores = det_scores[det_labels == c]  # (n_class_detections)
        n_class_detections = det_class_boxes.size(0)
        if n_class_detections == 0:
            continue

        # Sort detections in decreasing order of confidence/scores
        det_class_scores, sort_ind = torch.sort(det_class_scores, dim=0, descending=True)  # (n_class_detections)
        det_class_images = det_class_images[sort_ind]  # (n_class_detections)
        det_class_boxes = det_class_boxes[sort_ind]  # (n_class_detections, 4)

        # In the order of decreasing scores, check if true or false positive
        true_positives = torch.zeros((n_class_detections), dtype=torch.float).to(device)  # (n_class_detections)
        false_positives = torch.zeros((n_class_detections), dtype=torch.float).to(device)  # (n_class_detections)
        # FOR AVERAGE JACCARD PER CLASS FOR TRUE POSITIVES!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
        jaccard_TP = torch.zeros((n_class_detections), dtype=torch.float).to(device)

        for d in range(n_class_detections):
            this_detection_box = det_class_boxes[d].unsqueeze(0)  # (1, 4)
            this_image = det_class_images[d]  # (), scalar

            # Find objects in the same image with this class, and whether they have been detected before
            object_boxes = true_class_boxes[true_class_images == this_image]  # (n_class_objects_in_img)
            # If no such object in this image, then the detection is a false positive
            if object_boxes.size(0) == 0:
                false_positives[d] = 1
                continue

            # Find maximum overlap of this detection with objects in this image of this class
            overlaps = find_jaccard_overlap(this_detection_box, object_boxes)  # (1, n_class_objects_in_img)
            max_overlap, ind = torch.max(overlaps.squeeze(0), dim=0)  # (), () - scalars

            # 'ind' is the index of the object in these image-level tensors 'object_boxes', 'object_difficulties'
            # In the original class-level tensors 'true_class_boxes', etc., 'ind' corresponds to object with index...
            original_ind = torch.LongTensor(range(true_class_boxes.size(0)))[true_class_images == this_image][ind]
            # We need 'original_ind' to update 'true_class_boxes_detected'

            # If the maximum overlap is greater than the threshold of 0.5, it's a match
            if max_overlap.item() > 0.5:
                # If this object has already not been detected, it's a true positive
                if true_class_boxes_detected[original_ind] == 0:
                    true_positives[d] = 1
                    # ADDING OVERLAP IF TP!!!!!
                    jaccard_TP[d] = max_overlap
                    true_class_boxes_detected[original_ind] = 1  # this object has now been detected/accounted for
                # Otherwise, it's a false positive (since this object is already accounted for)
                else:
                    false_positives[d] = 1
            # Otherwise, the detection occurs in a different location than the actual object, and is a false positive
            else:
                false_positives[d] = 1


        # JACCARD_TP AVERAGE FOR TP ONLY (THROWING OUT NON-ZERO ENTRIES)!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
        if len(jaccard_TP[jaccard_TP > 0]) != 0:
            jaccard_tp_average[c - 1] = jaccard_TP[jaccard_TP > 0].mean().item()
        else:
            jaccard_tp_average[c - 1] = 0.0
        
        # CONVENTIONAL PRECISION, RECALL AND F1-SCORES FOR THE CLASS (0.5 THRESHOLD)!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
        simple_precisions[c - 1] = true_positives.sum().item() / (true_positives.sum().item() + false_positives.sum().item() + 1e-8)
        simple_recalls[c - 1] = true_positives.sum().item() / n_class_objects
        simple_f1_scores[c - 1] = 2 * simple_precisions[c - 1] * simple_recalls[c - 1] / (simple_precisions[c - 1] + simple_recalls[c - 1] + 1e-8)

        # Compute cumulative precision and recall at each detection in the order of decreasing scores
        cumul_true_positives = torch.cumsum(true_positives, dim=0)  # (n_class_detections)
        cumul_false_positives = torch.cumsum(false_positives, dim=0)  # (n_class_detections)
        cumul_precision = cumul_true_positives / (
                cumul_true_positives + cumul_false_positives + 1e-10)  # (n_class_detections)
        cumul_recall = cumul_true_positives / n_class_objects  # (n_class_detections)

        # Find the mean of the maximum of the precisions corresponding to recalls above the threshold 't'
        recall_thresholds = torch.arange(start=0, end=1.1, step=.1).tolist()  # (11)
        precisions = torch.zeros((len(recall_thresholds)), dtype=torch.float).to(device)  # (11)
        for i, t in enumerate(recall_thresholds):
            recalls_above_t = cumul_recall >= t
            if recalls_above_t.any():
                precisions[i] = cumul_precision[recalls_above_t].max()
            else:
                precisions[i] = 0.
        average_precisions[c - 1] = precisions.mean()  # c is in [1, n_classes - 1]!!!!!!!!!!!!!!!!!!!!!!!!!!!!!! NO BACKGROUND!!!!

    # Calculate Mean Average Precision (mAP)
    mean_average_precision = round(average_precisions.mean().item(), 3)

    # Keep class-wise average precisions in a dictionary
    # STARTING FROM 1 AS 0 IS THE INDEX OF THE BACKGROUND!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    average_precisions = {classes[c + 1]: round(v, 3) for c, v in enumerate(average_precisions.tolist())}
    jaccard_tp_average = {classes[c + 1]: round(v, 3) for c, v in enumerate(jaccard_tp_average.tolist())}
    simple_precisions = {classes[c + 1]: round(v, 3) for c, v in enumerate(simple_precisions.tolist())}
    simple_recalls = {classes[c + 1]: round(v, 3) for c, v in enumerate(simple_recalls.tolist())}
    simple_f1_scores = {classes[c + 1]: round(v, 3) for c, v in enumerate(simple_f1_scores.tolist())}

    return average_precisions, jaccard_tp_average, simple_precisions, simple_recalls, simple_f1_scores, mean_average_precision
```

```python
# FINAL FUNCTION!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
# INPUT: THE DATALOADER (VALIDATION ONE) AND THE MODEL!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
# OUTPUT: ALL THE METRICS CALCULATED: PRECISION, RECALL, F1-SCORE, 
    # AVERAGE JACCARD INDEX FOR TRUE POSITIVES (FOR EACH CLASS SEPARATELY) AND MAP!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
# ALL METRICS ARE CALCULATED ON THE WHOLE DATALOADER (VALIDATION ONE)!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
def evaluator(test_loader, model): 
    # Setting the model to the eval mode
    model.eval()

    # list to store true and detected boxes, labels and scores
    det_boxes = list()
    det_labels = list()
    det_scores = list()
    true_boxes = list()
    true_labels = list()

    with torch.no_grad():
        for batch in tqdm(test_loader):

            transform1 = transforms.Compose([
                transforms.Normalize(mean_per_channel, std_per_channel) 
            ])

            images_batch = transform1(batch['image'].to(device))
            boxes_batch = batch['bboxes']
            labels_batch = batch['label']
            
            # predicting location and scores for each prior!!!!!
            predicted_locs, predicted_scores = model(images_batch)

            # filtering priors, so only boxes are left which detect some object rather than background 
                # (filters are set manually of course (hyperparameters kind of))
                # SO THESE ARE THE FINAL PREDICTIONS OF THE MODEL!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
            # OF COURSE, HYPERPARAMETERS CAN BE CHANGED, OR EVEN OPTIMAL HYPERPARAMETERS 
                # CAN BE FOUND, IF MAKE ANOTHER SPLIT OF DATA!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
            # PREDICTIONS ARE IN XY FORMAT (XMIN, YMIN, XMAX, YMAX)!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
            det_boxes_batch, det_labels_batch, det_scores_batch = model.detect_objects(predicted_locs, predicted_scores,
                                                                                       min_score = 0.25, max_overlap = 0.85,
                                                                                       top_k = 5)
            
            # Store this batch's results for calculation of mAP and other metrics!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
            boxes = [b.to(device) for b in boxes_batch]
            labels = [l.to(device) for l in labels_batch]

            # SINCE ALREADY IN XY FORMAT, I WILL CLAMP SUCH PREDICTIONS, SO THAT BOXES ARE INSIDE PHOTO!!!!!!
            for i in range(len(det_boxes_batch)):
                det_boxes_batch[i] = det_boxes_batch[i].clamp(0, 1)
            
            # JOINING (EXTENDING LIST) DETECTIONS AND TRUE VALUES FOR THE WHOLE DATA TOGETHER!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
            det_boxes.extend(det_boxes_batch)
            det_labels.extend(det_labels_batch)
            det_scores.extend(det_scores_batch)
            true_boxes.extend(boxes)
            true_labels.extend(labels)

        # CALCULATING ALL METRICS CONSIDERING THE WHOLE TEST(VALIDATION) DATASET!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
        # (AP for each class, average jaccard index for all True Positives per Class, precision per Class, 
            # recall per Class, average F1-score per Class, MAP)
        ap, jac_tp_av, sPr, sRe, sF1, map = calculate_mAP_PCavJTP_PCPrRe(det_boxes, det_labels, det_scores,
                                                                         true_boxes, true_labels)

        # returning all the metrics
        return ap, jac_tp_av, sPr, sRe, sF1, map
```
